```python
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain_community.document_loaders import PyPDFLoader,TextLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
pdf_loader = PyPDFLoader("Submission.pdf")
pdf=pdf_loader.load()
text_loader = TextLoader("sql_prompt.txt")
text=text_loader.load()
```

```python
splitter= RecursiveCharacterTextSplitter(chunk_size=50,chunk_overlap=20)
```

```python
text_chunks = splitter.split_documents(text)
print(len(text_chunks))
```

```python
pdf_chunks = splitter.split_documents(pdf)
print(len(pdf_chunks))
```

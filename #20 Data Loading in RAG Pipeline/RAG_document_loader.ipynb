### Text Documents Loader

```python
from langchain_community.document_loaders import TextLoader
loader = TextLoader("sql_prompt.txt")
text = loader.load()
len(text)# return the how many documents or pages in list.

print(text[0].page_content)
```

### PDF Document Loader

```python
from langchain_community.document_loaders import PyPDFLoader
loader = PyPDFLoader("Submission.pdf")
pdf = loader.load()
len(pdf) # return list of pages or documents.

print(pdf[0].page_content)
```

### CSV loader

```python
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain_core.tools import tool

@tool
def addition_tool(a:int, b:int):
    """
    this function will return addition on two numbers.
        arge :
            a=number one
            b=number two
    """
    return a + b
    
```

```python
addition_tool.invoke({"a":12,"b":13})
```

```python
from langchain_openai import ChatOpenAI
from langchain.agents import create_agent
```

```python
llm = ChatOpenAI(model="gpt-4")
```

```python
agent = create_agent(
    model=llm,
    tools=[addition_tool],
    system_prompt="you are a math teacher and always use tools for calcuations."
)
response = agent.invoke({"messages":[{"role":"user","content":"what is the result of 2 + 2?"}]})
```

```python
response
```

```python
for res in response["messages"]:
    print(res)
    print("\n")
```

### Output  Structure for llm


```python
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain_openai import ChatOpenAI
from pydantic import BaseModel, Field
from typing import List # this is use to define list inside the output_structure we dont have str like word for list do we use this to create list. 
llm = ChatOpenAI(model="gpt-4o")
```

```python
class OutputStructure(BaseModel):
    name:str = Field(description="Full name")
    age:int= Field(description="age")
    email:str=Field(description="email")

class Allmovies(BaseModel):
    movies:List[movies] 

llm_response_Structure = llm.with_structured_output(OutputStructure)
res=llm_response_Structure.invoke(f"my name is Nikunj Shetye my age is 27 and my email is nikunj.shetye@gmail.com extact information from this text")
```

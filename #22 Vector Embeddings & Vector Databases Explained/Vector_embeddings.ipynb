```python
from dotenv import load_dotenv
load_dotenv()
```

```python
documents = [
    "The quick brown fox jumps over the lazy dog.",
    "Machine learning models require large amounts of training data.",
    "The stock market experienced significant volatility this quarter.",
    "She walked along the beach collecting seashells at sunset.",
    "Distributed systems must handle network partitions gracefully.",
    "The chef added a pinch of saffron to enhance the flavor.",
    "Climate change is causing more frequent extreme weather events.",
    "He spent the weekend hiking through the national park.",
    "The new smartphone features a faster processor and better camera.",
    "Researchers discovered a new species of frog in the rainforest.",
    "The concert was postponed due to heavy rainfall.",
    "Investing early can significantly boost long-term retirement savings.",
    "The museum's new exhibit showcases ancient Egyptian artifacts.",
    "Regular exercise improves both physical and mental health.",
    "The startup secured a major round of venture capital funding.",
    "Astronomers observed a rare alignment of five planets.",
    "The novel explores themes of identity and belonging.",
    "Traffic congestion in the city has worsened over the past decade.",
    "The bakery's fresh sourdough sells out within an hour of opening.",
    "Engineers are developing more efficient batteries for electric vehicles.",
]
```

```python
from langchain_openai import OpenAIEmbeddings
```

```python
embeddings = OpenAIEmbeddings(model="text-embedding-3-large")
```

```python
vector = embeddings.embed_documents(documents)
```

```python
print(len(vector[0]))
```

#### Store vector embeddings in vector DB(local)

```python
from langchain_chroma import Chroma
```

```python
vector_stored = Chroma.from_texts(
    texts=documents,
    embedding=embeddings,
    persist_directory="./vector_db"
)
```

### see how to fetch besed on the user requirement.

```python
query ="machine learning"
result =vector_stored.similarity_search(query=query,k=2) # k is the limit, meaning this will show the closest 2 based on the query
result
```
